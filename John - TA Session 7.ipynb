Below is a block of code that cleans the data frame so that the incomes are numeric, any row with NaN's is dropped, and then the Rank is reordered.

Now that was a lot of work, _and_ a lot of code! I don't want to have to repeat that process for evey combination of years, so let's generalize with a function!

```python
#I set yr2 to an empty pandas series, as that is what yr 2 will be.
def income_plots(yr1, label1, yr2=pd.Series(0), label2=False):
    fig, ax = plt.subplots()
    ax.plot(rank, yr1, 'g.-', label=label1)
    #Because yr2 is a series, I must use .any() to indicate that if any aspect of yr2 evaluates to true (non zero) proceed
    if yr2.any():
        ax.plot(rank, yr2, 'r.-', label=label2)
    ax.set_ylabel('Average Income')
    ax.set_xlabel('State Rank')
    if yr2.any():
        ax.set_title('Average Income by State Ordered By Rank')
    else: 
        ax.set_title(f'Average Income by State Ordered By Rank ({label1})')
    #note that vline indicates a vertical line and hline indicates a horizontal line
    ax.axvline(25, color='k', linestyle=':')
    ax.axhline(yr1[24], color='k', linestyle=':')
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    #loc is how you decide where the legend goes, best allows matplotlib to decide
    ax.legend(loc='best')    
    plt.show()
```

Let's try it out!

```python
income_plots(yr_2018, '2018')
```

```python
income_plots(yr_2018, '2018', yr_2014, '2014')
```

```python
income_plots(yr_2015, '2015', yr_2016, '2016')
```

```python
income_plots(yr_2017, '2017')
```

Now let's try one more experiment. First, let's transpose the data so that each state is a column. Then, let's make bar graphs to show how average income has changed within states.

```python
df = inc_df.copy()
df.index = df['State or territory']
df = df.drop(['State or territory', 'Rank'], axis=1)
#This is how to transpose a pandas dataframe
df = df.T
df.reset_index(inplace=True)
df = df.rename(columns={'index':'Year'})
#Because Year came from the index, it is an object, it isn't necessary, but let's turn it into a numeric
df['Year'] = pd.to_numeric(df['Year'])
df
```

Alright, with the reshaping out of the way, let's make some bar graphs!

You'll notice that the difference is we add .bar to the end of .plot, there is a similar convention for other types of plots as well (.pie is a pie chart for example). 

```python
ax = df.plot.bar(x='Year', y='Washington, D.C.', color='g')
```

As you can see, there are some problems with this graph. We don't need a legend, but we do need a title! Also, the bars are in descending order. Let's fix that!

```python
df = df.sort_values(by=['Year'])
ax = df.plot.bar(x='Year', y='Washington, D.C.', color='g', legend=None)
ax.set_title('Average Income in D.C. Over Time')
plt.show()
```

It's obvious that the y-axis is years, so let's remove that label. Let's also rotate the years so that you don't have to tilt your head to read them. Let's also add $ to the y label.

```python
ax = df.plot.bar(x='Year', y='Washington, D.C.', 
                 color='g', legend=None, rot=0)
ax.set_title('Average Income in D.C. Over Time')
#cite:https://www.markhneedham.com/blog/2018/09/18/matplotlib-remove-axis-legend/
ax.xaxis.set_label_text('')
#cite: https://www.codegrepper.com/code-examples/python/matplotlib+dollar+sign
ax.yaxis.set_major_formatter(ticker.FormatStrFormatter('$%1.0f'))
plt.show()
```

Finally, let's wrap all that code in a function, so we can explore any state!

```python
def state_barplt(state):
    ax = df.plot.bar(x='Year', y=state, 
                     color='g', legend=None, rot=0)
    ax.set_title(f'Average Income in {state} Over Time')
    #cite:https://www.markhneedham.com/blog/2018/09/18/matplotlib-remove-axis-legend/
    ax.xaxis.set_label_text('')
    #cite: https://www.codegrepper.com/code-examples/python/matplotlib+dollar+sign
    ax.yaxis.set_major_formatter(ticker.FormatStrFormatter('$%1.0f'))
    plt.show()
```
